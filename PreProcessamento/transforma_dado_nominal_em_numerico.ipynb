# Transformando dados categóricos nominais em numéricos

##  Usando sklearn.preprocessing.LabelEncoder

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
dados = pd.read_csv("dados.csv")
dados
```

```python
label_encoder = LabelEncoder()
features = ['MES', 'DIA_SEMANA']
for f in features:
    dados[f] = label_encoder.fit_transform(dados[f])
dados    
```

## Usando pandas.get_dummies()

```python
import pandas as pd
```

```python
dados = pd.read_csv("dados.csv")
dados
```

```python
mes = pd.get_dummies(dados['MES'])
mes
```

```python
dia_da_semana = pd.get_dummies(dados['DIA_SEMANA'])
dia_da_semana
```

```python
dados_dummy = pd.concat([dados, mes, dia_da_semana], axis = 1)
dados_dummy
```
